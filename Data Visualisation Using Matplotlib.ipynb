### Line Charts
#### Lets plot a graph for Web developer salary with respect to their age

```python
dev_x=np.arange(25,35)
dev_y=[37000,43000,45000,52000,56000,55000,60000,72000,68000,80000]
```

```python
plt.plot(dev_x,dev_y)
```

```python
#Adding Title , X,Y axis names
plt.plot(dev_x,dev_y)
plt.xlabel('Age')
plt.ylabel('Average Salary in USD')
plt.title("Average Salary vs Age")
```

```python
#Lets assume Python developer Salary and plot along with developers salary and provide Legends
py_dev_y=[39000,45000,47000,54000,58000,57000,63000,75000,70000,83000]
plt.plot(dev_x,dev_y,marker='.',label="All Developers")
plt.plot(dev_x,py_dev_y,marker='o',label="Python Developers")
plt.xlabel('Age')
plt.ylabel('Average Salary in USD')
plt.title("Average Salary vs Age")
plt.legend()
```

### Bar Charts

```python
dev_x=np.arange(25,35)
dev_y=[37000,43000,45000,52000,56000,55000,60000,72000,68000,80000]
plt.bar(dev_x,dev_y)
plt.xlabel('Age')
plt.ylabel('Average Salary in USD')
plt.title("Average Salary vs Age")
```

```python
py_dev_y=[39000,45000,47000,54000,58000,57000,63000,75000,70000,83000]
plt.bar(dev_x,dev_y,label="All Developers")
plt.bar(dev_x,py_dev_y,label="Python Developers")
plt.xlabel('Age')
plt.ylabel('Average Salary in USD')
plt.title("Average Salary vs Age")
plt.legend()
```

In the above chart we can see that Python dev values are overlapping on All Dev values, so in order to seperate the values for different developers do as shown below.

```python
x_index=np.arange(len(dev_x))
py_dev_y=[39000,45000,47000,54000,58000,57000,63000,75000,70000,83000]
plt.bar(x_index,dev_y,width=0.3,label="All Developers")
plt.bar(x_index+0.3,py_dev_y,width=0.3,label="Python Developers")
plt.xlabel('Age')
plt.ylabel('Average Salary in USD')
plt.title("Average Salary vs Age")
plt.legend()
```

In this plot you can see diff Developers values are seperated in Bar chart but if see x axis Age is not correct, so in order to set X axis with age value do as below

```python
x_index=np.arange(len(dev_x))
py_dev_y=[39000,45000,47000,54000,58000,57000,63000,75000,70000,83000]
plt.bar(x_index,dev_y,width=0.3,label="All Developers")
plt.bar(x_index+0.3,py_dev_y,width=0.3,label="Python Developers")
plt.legend()
plt.xticks(ticks=x_index,label=dev_x)
plt.xlabel('Age')
plt.ylabel('Average Salary in USD')
plt.title("Average Salary vs Age")
```

### Apply the Bar Chart to a CSV file which contains Data about programming Languages

```python
import csv
import numpy as np
import pandas as pd
from pandas import DataFrame
from collections import Counter
from matplotlib import pyplot as plt
```

```python
bar_chart_data=pd.read_csv('Bar Chart Data.csv')
bar_chart_data.head()
language_count=Counter()
for i in bar_chart_data['LanguagesWorkedWith']:
    language_count.update(i.split(';'))

#print(language_count)

## In order to find the most common languages used we use most_common() function, the below function selects top 15 programming languages

print(language_count.most_common(15))
```

Most_common(15) funtion returns 15 most commonly used programming languages.
observe the output is list containing tuples
So in order to seperate the languages and respective count we need to loop over the above output as below

```python
Languages=[]
Popularity=[]

for item in language_count.most_common(15):
    Languages.append(item[0])
    Popularity.append(item[1])


print(Languages)
print(Popularity)
```

Now lets plot the Bar graph for the above lists

```python
plt.barh(Languages,Popularity)##Horizontal Bar Chart
plt.title("Most Popular Programming Languages")
#plt.ylabel('Programming Language')
plt.xlabel('Number of People Using')
```

Here you can see 'JavaScript' which most users are using is at the bottom, in order to bring it on the top just reverse the List which we are passing earlier ie:: Language.reverse() and same Popularity.reverse()
